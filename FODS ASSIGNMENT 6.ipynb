```python
# Replace 0 values in RestingBP and Cholesterol with NaN for cleaning (as 0 is invalid for these fields)
data_cleaned = df.copy()
data_cleaned['RestingBP'] = data_cleaned['RestingBP'].replace(0, np.nan)
data_cleaned['Cholesterol'] = data_cleaned['Cholesterol'].replace(0, np.nan)
```

```python
# Handle negative values in Oldpeak by replacing them with NaN
data_cleaned['Oldpeak'] = data_cleaned['Oldpeak'].apply(lambda x: np.nan if x < 0 else x)
```

```python
# Impute missing values in RestingBP, Cholesterol, and Oldpeak with their respective medians
data_cleaned['RestingBP'] = data_cleaned['RestingBP'].fillna(data_cleaned['RestingBP'].median())
data_cleaned['Cholesterol'] = data_cleaned['Cholesterol'].fillna(data_cleaned['Cholesterol'].median())
data_cleaned['Oldpeak'] = data_cleaned['Oldpeak'].fillna(data_cleaned['Oldpeak'].median())

# Verify that there are no missing values after imputation
print("\nMissing Values After Imputation:\n", data_cleaned.isna().sum())
```

```python
# Step 4: Saving the cleaned dataset
data_cleaned.to_csv('cleaned_heart_disease_data.csv', index=False)
print("\nCleaned dataset saved as 'cleaned_heart_disease_data.csv'")
```
