```python
import pandas as pd
```

```python
# Load the dataset
df=pd.read_csv('/content/Air_Quality.csv')
```

```python
# Display the first few rows to check the data
df.describe()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Check for duplicate rows
df.duplicated()
```

```python
# Convert the 'Start_Date' to datetime format
df['Start_Date'] = pd.to_datetime(df['Start_Date'], format='%m/%d/%Y')
```

```python
df.describe()
```

```python

# Create a new 'Year' column from the 'Start_Date'
df['Year'] = df['Start_Date'].dt.year
```

```python
# Display the transformed data
print("\nTransformed Data:")
df.head()
```

```python
# Grouping data by 'Geo Place Name' and calculating the mean of 'Data Value' for each region
region_group = df.groupby('Geo Place Name')['Data Value'].mean().reset_index()
```

```python

# Display the grouped data
print("\nGrouped Data by Region:")
region_group.head()
```
